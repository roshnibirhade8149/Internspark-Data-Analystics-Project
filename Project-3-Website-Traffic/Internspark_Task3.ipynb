# DATA CLEANING

```python
import pandas as pd

df = pd.read_csv("traffic.csv")

# Convert date
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Remove duplicates
df = df.drop_duplicates()

# Check missing values
print(df.isnull().sum())
```

# KPIs

```python
# 1. Total Events

total_events = len(df)
print("Total Events:", total_events)
```

```python
# 2. Event Distribution

event_counts = df['event'].value_counts()
print(event_counts)
```

```python
# 3. CLICK-THROUGH RATE

clicks = event_counts['click']
views = event_counts['pageview']

ctr = (clicks / views) * 100
print("Click Through Rate:", ctr)
```

```python
# 4. Bounce Rate

link_counts = df.groupby('linkid').size()

bounce_links = link_counts[link_counts == 1].count()
total_links = df['linkid'].nunique()

bounce_rate = (bounce_links / total_links) * 100
print("Bounce Rate:", bounce_rate)
```

Data Limitation

The dataset does not contain user-level or session-level identifiers. Therefore, metrics such as total users, sessions, and average session duration could not be calculated. The analysis is based on event-level and link-level data.
